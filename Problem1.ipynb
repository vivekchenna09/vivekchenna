#**1. Read the file alice.txt as a full string (look at read_text_file.ipynb for how to read a text file line by line). Process it using NLTK library**

```python
# Using a deferent fuction for reading text file for read_all_lines function to be used for universal for any text input
def read_textfile():
  f = open('alice.txt', 'r') # opening text file
  all_lines = ' '.join(f.readlines()) #reading all lines in a textfile
  return all_lines # returning the data
```

**(a) Lower case all characters, tokenize all sentences, and extract all tokens:**

```python
from nltk import sent_tokenize, word_tokenize
# assigned text from the above function read_textfile to a variable called all_lines
all_lines=read_textfile()
text1 = all_lines
sents = sent_tokenize(text1)
tokens = [word_tokenize(sent.lower()) for sent in sents]

# Combine tokens from all sentences into a single list
all_tokens = [token for sent_tokens in tokens for token in sent_tokens]

# Print the total number of terms
print("Total number of terms:", len(all_tokens))

# Extract the list of unique tokens and their frequencies
from collections import Counter
term_frequencies = Counter(all_tokens)

# Sort the dictionary by frequencies from largest to smallest
sorted_term_frequencies = dict(sorted(term_frequencies.items(), key=lambda x: x[1], reverse=True))

# Print the most frequent 20 terms and their frequencies
print("\nMost frequent 20 terms:")
for term, frequency in list(sorted_term_frequencies.items())[:20]:
    print(f"{term}: {frequency}")

# Print the least frequent 20 terms and their frequencies
print("\nLeast frequent 20 terms:")
for term, frequency in list(sorted_term_frequencies.items())[-20:]:
    print(f"{term}: {frequency}")
```

**(b) Lemmatize words and repeat all steps from (a)**

```python
from nltk.stem import WordNetLemmatizer

# Initialize the lemmatizer
lemmatizer = WordNetLemmatizer()

# Lemmatize each token
lemmatized_tokens = [lemmatizer.lemmatize(token) for token in all_tokens]

# Extract the list of unique lemmatized tokens and their frequencies
lemmatized_term_frequencies = Counter(lemmatized_tokens)

# Sort the lemmatized dictionary by frequencies from largest to smallest
sorted_lemmatized_term_frequencies = dict(sorted(lemmatized_term_frequencies.items(), key=lambda x: x[1], reverse=True))

# Print the total number of lemmatized terms
print("\nTotal number of lemmatized terms:", len(lemmatized_tokens))

# Print the most frequent 20 lemmatized terms and their frequencies
print("\nMost frequent 20 lemmatized terms:")
for term, frequency in list(sorted_lemmatized_term_frequencies.items())[:20]:
    print(f"{term}: {frequency}")

# Print the least frequent 20 lemmatized terms and their frequencies
print("\nLeast frequent 20 lemmatized terms:")
for term, frequency in list(sorted_lemmatized_term_frequencies.items())[-20:]:
    print(f"{term}: {frequency}")
```

**(c) Remove stop words and punctuation, and repeat all steps from (a). Additionally, plot the 50 largest frequencies:**

```python
from nltk.corpus import stopwords
from string import punctuation

# Define a function to remove stop words and punctuation
def remove_stopwords_and_punctuation(tokens):
    stop_words = set(stopwords.words("english"))
    return [token for token in tokens if token not in stop_words and token not in punctuation]

# Remove stop words and punctuation
filtered_tokens = remove_stopwords_and_punctuation(all_tokens)

# Extract the list of unique filtered tokens and their frequencies
filtered_term_frequencies = Counter(filtered_tokens)

# Sort the filtered dictionary by frequencies from largest to smallest
sorted_filtered_term_frequencies = dict(sorted(filtered_term_frequencies.items(), key=lambda x: x[1], reverse=True))

# Print the total number of filtered terms
print("\nTotal number of filtered terms:", len(filtered_tokens))

# Print the most frequent 20 filtered terms and their frequencies
print("\nMost frequent 20 filtered terms:")
for term, frequency in list(sorted_filtered_term_frequencies.items())[:20]:
    print(f"{term}: {frequency}")

# Print the least frequent 20 filtered terms and their frequencies
print("\nLeast frequent 20 filtered terms:")
for term, frequency in list(sorted_filtered_term_frequencies.items())[-20:]:
    print(f"{term}: {frequency}")

# Plot the 50 largest frequencies
max_length = 50
list_words = list(sorted_filtered_term_frequencies.keys())[:max_length]
list_words_chopped = [w[:2] for w in list_words]
plt.plot(list(sorted_filtered_term_frequencies.values())[:max_length], '.')
plt.xticks(ticks=range(min(max_length, len(sorted_filtered_term_frequencies))), labels=list_words, rotation=90)
plt.show()
```
